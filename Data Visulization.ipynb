## Line Chart

### A line chart is a chart that display informations as a series of data ponts called"markers" connected by staright line

```python
import matplotlib.pyplot as plt
week=[1,2,3,4]
Prices=[40,80,100,50]
plt.plot(week,Prices)
plt.xlabel('week')
plt.ylabel('Onion Prices (Rs.)')
plt.show()
```

```python
# Specifying Plot size and grid
x=[3,4,5,7,89,7]
y=[9,5,6,3,78,5]
plt.plot(figsize=(15,7)) # figure size specified 15 units wide x coordinates and 7 units wide y coordinates
plt.plot(x,'b') # 'b' indicates blue color
plt.plot(y,'r',linestyle="dashed")#  'r' for red and dashed line style
plt.xlabel('Xvalues')
plt.ylabel('yvalues')
plt.show()
```

### Scatter Plot

```python
# Scatter chart
x=[3,4,5,7,9,7]
y=[9,5,6,3,7,5]
plt.scatter(x,y)
plt.scatter(x,y,marker="x") # specifying the shape of the marker
plt.scatter(x,y,s=9,c='m',marker="x") # s specify the size of the marker, c specify the color code
```

```python
#Scatter Plot
import numpy as np
x=np.random.randint(1,100,size=(100,))
y=np.random.randint(1,100,size=(100,))
plt.scatter(x,y,color='r')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()                
```

#### Bar charts is a graphical display of using bars of different heights to represent categorical data , they can be horizontally and vertically

### Bar Chart

```python
# Bar charts
Cities=['Delhi','Noida','Mumbai','Hyderabad']
Populations=[234586349,3784939047,36494964903,12634074648]
plt.bar(Cities,Populations)
plt.xlabel('Cities') # specifying the label for x axis
plt.ylabel('Populations') # specifying the label for y axis
plt.show()
```

```python
# Bar chart example
# Plot medals won by India and Australia both in one chart
Info=['Gold','Silver','Bronze','Total']
Aust=[80,59,59,198]
Ind=[26,20,20,66]
plt.bar(Info,Aust,width=[0.7,0.5,0.3,1])# we can specify the width of the grid
plt.bar(Info,Ind,width=[0.7,0.5,0.3,1])
plt.xlabel("Medal Types")
plt.ylabel("Totals medals")
plt.show()
```

```python
# Bar chart example changing colors of the bars
# Plot medals won by India and Australia both in one chart
Info=['Gold','Silver','Bronze','Total']
Aust=[80,59,59,198]
Ind=[26,20,20,66]
plt.bar(Info,Aust,width=[0.7,0.5,0.3,1],color=['gold','Black','Green','yellow'])# we can specify the width of the grid
plt.bar(Info,Ind,width=[0.7,0.5,0.3,1])
plt.xlabel("Medal Types")
plt.ylabel("Totals medals")
plt.show()
```

### Pie chart

#### A Pie chart is a circular statstic chart , which is divided inti slices to illustrate the numercal proportion

```python
divide=[30,56,20,10,89]
plt.pie(divide)
```

```python
divide=[30,56,20,10,89]
houses=['Rita','Vidya','Namrata','Aditi','Sakshi']
plt.pie(divide,labels=houses) # labels from houses will be taken in sequence
```

#### To view percentage of share in a pie chart , you need to add an argument "autopct" with aformat string , such as "%1.1F%%"

```python
divide=[30,56,20,10,89]
houses=['Rita','Vidya','Namrata','Aditi','Sakshi']
plt.pie(divide,labels=houses,autopct="%1.1f%%")
```

```python
# Changing colors of the pie
colr=['Red','blue','black','yellow']
divide=[20,50,45,10]
plt.pie(divide,colors=colr,labels=colr)
```

```python
# Adding title in a chart and legend
import numpy as np
col1=[800,1000,9800,7300,11200]
col2=[50,68,98,45,78]
plt.plot(col1,color='b',label='range1')
plt.plot(col2,color='r',label='range2')
plt.legend(loc='upper left') # setting the position for legends
plt.title('col1 and col2 visulization') # setting the title of the chart
plt.xlabel('col1')
plt.ylabel('col2')
plt.show()


```

## Box Plot

### It is a representation of the stastical techniques for presenting the 5 - number summary which consists of:
#### 1. Minimum range
#### 2. Max Range
#### 3. the upper quartile
#### 4. the lower quartile
#### 5. the median

```python
ary=[5,20,30,45,60,80,100,140,150,200,240]
plt.boxplot(ary)
plt.boxplot(ary,showmeans=True) # shows the mean value
```

### Box plot helps us to identify the outliers in our data
